```python
import sympy as sp
import matplotlib.pyplot as plt
import numpy as np
```

## 1. Calcule

```python
# a) lim x->0 sin(2x)/x

x = sp.Symbol('x')
f = sp.sin(2*x)/x
f_deriv = sp.diff(f, x)


print(f"Limite: {sp.limit(f,x,0)}, Derivada: {f_deriv}")
```

```python
x_vals = np.linspace(-10, 10, 500)
y_vals = [f.subs(x,i) for i in x_vals]

fig, ax = plt.subplots()
ax.plot(x_vals, y_vals, color="green")
plt.xlabel('x')
plt.ylabel('lim(x)')
ax.grid()
fig.show
```

```python
#b) lim x-> âˆž (3x^4 + 6x^3 - 5x^2 + 6x - 1)/ 5x^2 + 5x - 8
x = sp.Symbol('x')
f = ((3*x**4) + (6*x**3) - (5*x**2) + 6*x - 1)/(5*x**2 + 5*x - 8)
f_deriv = sp.diff(f, x)


print(f"Limite: {sp.limit(f,x,np.inf)}, Derivada: {f_deriv}")
```

```python
x_vals = np.linspace(-10, 10, 500)
y_vals = [f.subs(x,i) for i in x_vals]
fig, ax = plt.subplots()
ax.plot(x_vals, y_vals, color="green")
plt.xlabel('x')
plt.ylabel('lim(x)')
ax.grid()
fig.show
```

```python
#c) lim x-> 0 (1+2x)^1/x

x = sp.Symbol('x')
f = (1+2*x)**(1/x)
f_deriv = sp.diff(f, x)


print(f"Limite: {sp.limit(f,x,0)}, Derivada: {f_deriv}")
```

```python
x_vals = np.linspace(0, 10, 500)
y_vals = [f.subs(x,i) for i in x_vals]
fig, ax = plt.subplots()
ax.plot(x_vals, y_vals, color="green")
plt.xlabel('x')
plt.ylabel('lim(x)')
ax.grid()
fig.show
```

```python
#d) lim x->2 (x^4-16)/(x-2)

x = sp.Symbol('x')
f = (x**4-16)/(x-2)
f_deriv = sp.diff(f, x)


print(f"Limite: {sp.limit(f,x,2)}, Derivada: {f_deriv}")
```

```python
#e) lim x-> 2+ 3/2-x

x = sp.Symbol('x')
f = 3/(2-x)
f_deriv = sp.diff(f, x)


print(f"Limite: {sp.limit(f,x,2,dir='+')}, Derivada: {f_deriv}")
```

### c) desenhe utilizando Python as funÃ§Ãµes S(t), v(t)) e a(t) em um Ãºnico sistema. 1


```python
t = np.linspace(-5, 5, 100)
S = -3*t**2 - 5*t + 6
v = -6*t -5
a = -6


plt.plot(t, S, label='S(t)', color="green")
plt.plot(t, v, label='v(t)', color="orange")
plt.axhline(y=a, color='blue', label='a(t)')

plt.xlabel('t')
plt.ylabel('S(t), v(t), a(t)')

plt.legend()
plt.grid()
plt.show()
```
